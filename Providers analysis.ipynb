```python
import pandas as pd

prov = pd.read_csv('providers.csv')
prov['Funding'] = prov['2014-15 Australian Government Funding'].fillna(0).apply(lambda x: int(str(x).replace('n.a.','0').replace('$','').replace(',','')))
prov.head()
```

```python
# Let's just look at residential care
resi = prov[(prov['Residential Places'].notnull())&(prov['Home Care Low Places'].isnull())\
            &(prov['Home Care High Places'].isnull())&(prov['Transition Care Places'].isnull())]
resi.sort_values('Residential Places')
```

```python
resi.groupby('Physical Address State')[['Residential Places','Home Care Low Places','Home Care High Places','Transition Care Places','Funding']]\
.sum().sort('Residential Places',ascending=False)
```

```python
resi['$pp'] = resi['Funding'] / resi['Residential Places']
resi.groupby(['Organisation Type','Physical Address State'])[['Residential Places','Funding','$pp']]\
.sum()
```

```python
resi['$pp'] = resi['Funding'] / resi['Residential Places']
resi.groupby(['Organisation Type'])[['Residential Places','Funding','$pp']]\
.sum().sort_values('$pp')
```
