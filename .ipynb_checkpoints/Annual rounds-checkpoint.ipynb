```python
import pandas as pd

four = pd.read_csv('tabula-2014.csv')
five = pd.read_csv('tabula-2015.csv')

def nums(x):
    final = ''
    for letter in str(x):
        try:
            final += final + str(int(letter))
        except:
            pass
    return int(final)
```

```python
import re
import numpy as np
four[['Place made','Applications received','Places sought']] = four[['Places made','Applications received','Places sought']]\
.applymap(lambda x: re.sub(r'[^\d.]+', '', str(x))).replace("",np.nan).astype(float)
five[['Place made','Applications received','Places sought']] = five[['Places made','Applications received','Places sought']]\
.applymap(lambda x: re.sub(r'[^\d.]+', '', str(x))).replace("",np.nan).astype(float)

four['Year'] = ['2014']*len(four)
five['Year'] = ['2015']*len(five)
```

```python
df = pd.concat([four,five])
df['Type of place '] = df['Type of place '].apply(lambda x: str(x).strip())
df
```

```python
df.groupby(['Type of place ','Area','Year']).sum()
```

```python
allo = pd.read_csv('allocations.csv')
allo['Places Allocated'] = allo['Places Allocated'].replace('-',np.nan).astype(float)
allo[' Capital Grant '] = allo[' Capital Grant '].apply(lambda x: str(x).replace("$",'').replace(",","")).replace('-',np.nan).astype(float)
```

```python
allo.columns
```

```python
allo2015 = allo.groupby(['State/\rTerritory','Region']).sum()
```

```python
five['Type of place '] = five['Type of place '].apply(lambda x: str(x).strip())
resiadv = five[five['Type of place ']=='Residential aged care places']
resiadv['Area'] = resiadv['Area'].apply(lambda x: str(x).strip())
allo2015 = allo2015.reset_index()
allo2015['Region'] = allo2015['Region'].apply(lambda x: str(x).strip())
len(resiadv)
```

```python


fiveallo = pd.merge(resiadv,allo2015,left_on=['Area','State'],right_on=['Region','State/\rTerritory'],how='outer')
print len(five)

print len(fiveallo)
fiveallo
```
